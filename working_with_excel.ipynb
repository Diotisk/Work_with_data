```python
import openpyxl
from openpyxl.utils.dataframe import dataframe_to_rows
import pandas as pd
```

```python
wb = openpyxl.load_workbook("/Users/daraeremina/Downloads/assignment.xlsx")
sheet_one = wb["Tabelle1"]
sheet_two = wb["Tabelle2"]
```

```python
df_one = pd.DataFrame(sheet_one.values)
df_two = pd.DataFrame(sheet_two.values)
df_one_col_one = df_one.get_value(0, 0)
df_one_col_two = df_one.get_value(0, 1)
df_two_col_one = df_one.get_value(0, 0)
df_two_col_two = df_one.get_value(0, 1)
```

```python
df_one_new = df_one.rename(columns={0: df_one_col_one, 1: df_one_col_two})
df_two_new = df_two.rename(columns={0: df_two_col_one, 1: df_two_col_two})
```

```python
all_data = pd.merge(df_one_new, df_two_new, on="Country")
result_data = all_data.drop(0).rename(columns={"Health expenditure and financing 2012, in Million US-Dollars_y": "Population"})
clean_result_data = result_data[result_data.Population != ".."]
clean_result_data
```

```python
def get_exp_per_inhab(x, y):
    return x / y

exp_per_inhab = list(clean_result_data.apply(lambda row: get_exp_per_inhab(row["Health expenditure and "
                                                              "financing 2012, "
                                                              "in Million US-Dollars_x"],
                                                          row["Population"]), axis=1))

clean_result_data.loc[:, "Health expenditure per inhabitant 2012, in Million US-Dollars"] = exp_per_inhab
```

```python
end_data = clean_result_data.filter(items=["Country", 
                                    "Health expenditure per inhabitant 2012, "
                                    "in Million US-Dollars"])
```

```python
my_wb = openpyxl.Workbook()
my_ws = my_wb.active

for r in dataframe_to_rows(end_data, index=False, header=True):
    my_ws.append(r)

my_wb.save("/Users/daraeremina/Downloads/assignment_result.xlsx")
```
